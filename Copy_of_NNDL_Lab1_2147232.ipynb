## Class For activation function

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
class Lab1:
  x=[]
  def __init__(self,str,df):
    self.x=df
    print( " This is Object to implement all the Activation Function  with the data "+str)

  def unitStep(self):
    df=[]
    for i in self.x:
      if i<0:
        df.append(0)
      elif i>0:
        df.append(1)
      else:
        df.append(0.5)
    self.make_plot(self.x,df,"unit step")

  def signum(self):
    df=[]
    for i in self.x:
      if i<0:
        df.append(-1)
      elif i>0:
        df.append(1)
      else:
        df.append(0)
    self.make_plot(self.x,df,"Sign")

  def linear(self):
    self.make_plot(self.x,self.x,"Linear")

  def pieceWiseLinear(self):
    df=[]
    for i in self.x:
      if i>=0.5:
        df.append(1)
      elif 0.5<i>-0.5:
        df.append(-0.5)
      else:
        df.append(0)
    self.make_plot(self.x,df,"Piece wise linear")

  def sigmoid(self):
    f = 1/(1 + np.exp(-(self.x)))
    self.make_plot(self.x, f, "Sigmoid")
    

  def hyperBolicTangent(self):
    f = (( np.exp(self.x))-np.exp(-(self.x)))/(( np.exp(self.x))+np.exp(-(self.x)))
    self.make_plot(self.x, f, "HyperBolic")
  
  # the auxiliary function forming the diagram
  def make_plot(self,x, f, name):
      plt.figure()
      plt.figure(figsize=(12,6))
      plt.title(name, fontsize=20, fontweight='bold')
      plt.xlabel('z', fontsize=15)
      plt.ylabel('Activation function value', fontsize=15)
      sns.set_style("whitegrid")
      plt.plot(x, f, label="f (z)")
      plt.legend(loc=4, prop={'size': 15}, frameon=True,shadow=True, facecolor="white", edgecolor="black")
      plt.show()
      


```
