```python
import numpy as np
import matplotlib.pyplot as plt
```

Sigmoidal Function

```python
x = np.arange(-10,10,1)
z = 1/(1+ np.exp(-x))

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("logistic (x)")
plt.grid()
plt.show()
```

Linear

```python
z = x

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("logistic (x)")
plt.grid()
plt.show()
```

Hyperbolic

```python
z = ((np.exp(x)-np.exp(-x)) / (np.exp(x)+np.exp(-x)))


plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("logistic (x)")
plt.title("Hyperbolic Tangent Function")
plt.grid()
plt.show()
```

```python
x = np.arange(-10,10,1)
y = []
for i in x:
   if i >= 0.5:
     y.append(1)
   elif -0.5<i<0.5:
     y.append(i+0.5)
   elif i<=-0.5:
     y.append(0)
  
plt.plot(x,y)
plt.grid()
plt.title("Peacewise Linear")
```

Sign Function

```python
x = np.arange(-10,20,0.2)
y = []
for i in x:
   if i < 0:
     y.append(-1)
   elif i == 0:
     y.append(0)
   elif i>0:
     y.append(1)
  
plt.plot(x,y)
plt.title("Sign Function")
plt.grid()
```

Unit Step Function

```python
x = np.arange(-20,10,0.5)
y = []
for i in x:
   if i < 0:
     y.append(0)
   elif i == 0:
     y.append(0.5)
   elif i>0:
     y.append(1)
  
plt.plot(x,y)
plt.title("Unit Step Function")
plt.grid()
```

### Activation functions with Iris dataset

```python
from sklearn import datasets
iris = datasets.load_iris()
import pandas as pd
```

```python
iris_df=pd.DataFrame(iris.data)
iris_df['class']=iris.target

iris_df.columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']
iris_df.dropna(how="all", inplace=True)
```

```python
iris_df.head()
```

```python
X = iris_df['sepal_len']
y = []
for i in X:
   if i < 0:
     y.append(0)
   elif i == 0:
     y.append(0.5)
   elif i>0:
     y.append(1)
  
plt.plot(X,y)
plt.grid()
plt.title("Unit Step Function")
```

```python
X = iris_df['sepal_len'] 
y = []
for i in X:
   if i < 0:
     y.append(-1)
   elif i == 0:
     y.append(0)
   elif i>0:
     y.append(1)
  
plt.plot(X,y)
plt.grid()
plt.title("Sign Function")
```

```python
X = iris_df['sepal_len'] 
plt.plot(X,X)
plt.grid()
plt.title("Linear Function")
```

```python
X = iris_df['petal_wid'] 
y = []
for i in X:
   if i >= 0.5:
     y.append(1)
   elif -0.5<i<0.5:
     y.append(i+0.5)
   elif i<=-0.5:
     y.append(0)
  
plt.plot(X,y)
plt.grid()
plt.title("Piecewise Linear")
```
