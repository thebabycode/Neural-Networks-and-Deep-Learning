```python
import numpy as np
import matplotlib.pyplot as plt
```

Sigmoidal Function

```python
x = np.arange(-10,10,1)
z = 1/(1+ np.exp(-x))

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("logistic (x)")

plt.show()
```

Linear

```python
z = x

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("logistic (x)")

plt.show()
```

Hyperbolic

```python
z = ((np.exp(x)-np.exp(-x)) / (np.exp(x)+np.exp(-x)))


plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("logistic (x)")

plt.show()
```

Piece Wise Linear

```python
z = if(x>=0.5,1, if(-0.5<x<0.5,x+0.5,if(x<=-0.5,0)))

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("logistic (x)")

plt.show()
```
